```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
# 1. seaborn의 flights 데이터셋을 로딩해서 시각화

fig = plt.figure(figsize = (12, 6), facecolor = 'aliceblue')

ax1 = fig.add_subplot(121, facecolor = 'seashell')
ax2 = fig.add_subplot(122, facecolor = 'ivory')

ax1_year = flights.groupby('year')
ax1_year_passengers = ax1_year['passengers'].sum()

ax1.bar(ax1_year_passengers.index, ax1_year_passengers, tick_label = ax1_year_passengers.index)
ax1.set_title('Total Passengers by Year', loc = 'center', fontsize = 15, fontweight = 'bold')

ax1.set_xlabel('Year', loc = 'right')
ax1.set_ylabel('Total Passengers', loc = 'top')

ax2_month = flights.groupby('month')
ax2_month_passengers = ax2_month['passengers'].sum()

ax2.plot(ax2_month_passengers.index, ax2_month_passengers, solid_joinstyle ='bevel', linewidth = 4)
ax2.set_title('Total Passengers by Month', loc = 'center', fontsize = 15, fontweight = 'bold')

ax2.set_xlabel('Month', loc = 'right')
ax2.set_ylabel('Total Passengers', loc = 'top')

plt.show()
```

```python
penguins = sns.load_dataset('penguins')
penguins
```

```python
fig = plt.figure(figsize = (10, 5))

ax = fig.add_subplot(111)

species = penguins['species'].unique()

# 산포도를 그릴 때 범례 표시하는 방법

for s in species  :
    x = penguins.loc[penguins['species'] == s, 'bill_length_mm']
    y = penguins.loc[penguins['species'] == s, 'bill_depth_mm']
    ax.scatter(x, y, alpha = 0.5, label = s)
    
ax.set_title("Penguins Dataset - Bill Length vs Bill Depth")
ax.set_xlabel('Bill Length (mm)')
ax.set_ylabel('Bill Depth (mm)')

ax.legend(loc = 'lower right', fontsize = 12)
plt.show()
```
