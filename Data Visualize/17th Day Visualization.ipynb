```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Text 그래프 중간에 text를 보여주기위한 용도

r = np.random.rand(5)
plt.figure()
plt.plot(r)

# x, y 위치에 text를 넣어주기위한 for문
for i, v in enumerate(r) : # enumerate 값을 반환하면서 index 값도 반환한다.
    plt.text(i, v, '{:.3f}'.format(v), ha = 'center', va = 'baseline') # 그래프와 글자의 위치 조절 좌표를 조금 변경하거나, 인자 사용 
                                                                       # horizontalalignment : 수평 위치, verticalalignment : 수직 위치, multialignment : 여러 줄 일때 

# Text가 화면을 벗어나는 경우 : 축을 조정한다.
plt.xlim(-0.5, 4.5) # 축 확장
plt.ylim(0, 1.2)
plt.show()
```

# box plot 데이터 분석에 중요한 형식

* 데이터 표본의 통계값들을 하나의 박스에 모아두는 형식

* 중앙값 : 데이터의 중앙위치의 값 ex.100개중 50번째 값

* 사분범위 : 3사번위수 - 1사분위수, IQR

* 잠재적 이상치 : Q3 + 1.5 * IQR Q1 - 1.5 * IQR의 범위 밖의 값 (통계상 희귀한 값, 정상 범위를 벗어난 값)

* 정상 범위 : 잠재적 이상치의 간격

```python
from sklearn.datasets import load_iris

iris = load_iris()

iris
X = iris['data']
X
```

```python
df = pd.DataFrame(X, columns = iris.feature_names)
```

```python
plt.figure()
plt.boxplot(df, labels = df.columns, notch = True, boxprops = { 'facecolor' : 'red'}, patch_artist = True) # 중앙값, 상위, 하위 20% 
plt.show()
```

```python
# 원형 그래프

# 원형 그래프에서 간격 띄우기, 크기 조절 (강조)
plt.pie([15, 10, 15], labels = ['Apple', 'Orange', 'Grape'], radius = 0.5,  # radius : 반지름으로 원 크기 조절
        explode = [0.02, 0.01, 0.01], colors = ['red', 'blue', 'green']) # explode : 반지름에서 떨어지는 정도

plt.show()
```

```python
# 데이터 분석하다가 추가로 더 그리는 경우
# 시계열 데이터에서 A라는 항목 중 특정 값이 넘는 곳을 확인하고싶다.
# 기준선을 그어서 명확히 확인한다?

r1 = np.random.rand(10)
r2 = np.random.rand(10)
```

```python
plt.figure()
plt.plot(r1)
plt.plot(r2)

plt.axhline(0.75, xmin = 0.3, xmax = 0.7, color = 'red') # axhline : axis horizontal line 축에대해 수평선 긋기  시계열에서 많이 쓴다.
plt.axvline(6, ymin = 0.1, ymax = 0.3, color = 'green') # axvline : axis vertical line 축에대해 수직선 긋기 특정 시간 이후 확인
                                                        # xmin, xmax, ymin, ymax 인자로 범위, 구간을 지정할 수 있다. 0 ~ 1 %로 지정
plt.axhspan(0.2, 0.4, xmin = 0, xmax = 0.7, alpha = 0.3, color = 'orange', hatch = '\\') # 그래프 구역 색칠, 테두리 없애기, 색 채우기 비우기 등등 다양한 인자가 존재
plt.axvspan(5, 7, ymin = 0.1, ymax = 1, alpha = 0.3, color = 'purple', hatch = '/', fill = False) # hatch : 무늬를 주는 인자, hatch 간격, 색, 테두리 색 등등 다양한 인자가 존재, 

# plt.fill_between(np.arange(2, 7), r1[2 : 7]) # 둘 사이를 채움 r1에 대해서 2 ~ 6 까지 색칠한다.
plt.fill_between(np.arange(2, 7), r1[2 : 7], r2[2 : 7], color = 'lime', alpha = 0.4) # 2 ~ 7에서 r1과 r2 사이를 칠해준다. 같아야 하는 데이터를 비교할 때 사용한다.

plt.show()
```
