# 막대 그래프

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

if os.name == 'posix' :
    plt.rc('font', family = 'AppleGothic')
else :
    plt.rc('font', family = 'Malgun Gothic')
```

```python
# 막대 그래프 형태 그리기 

fig = plt.figure()

ax = fig.add_subplot(111)

x = [1, 2]
y = [1, 3]

ax.bar(x, y)

plt.show()
```

```python
# 막대 그래프 X축 label 변경

fig = plt.figure()

ax = fig.add_subplot(111)

x = [1, 2]
y = [1, 3]

labels = ['Apple', 'Orange']
ax.bar(x, y, tick_label = labels) # tick_label : 막대 눈금의 이름 지정

plt.show()
```

```python
# 막대 그래프 Y축으로 그리기

fig = plt.figure()

ax = fig.add_subplot(111)

x = [1, 2]
y = [1, 3]

labels = ['Apple', 'Orange']
ax.barh(x, y, tick_label = labels) # barh : 막대 그래프의 x축 y축 변경 ?

plt.show()
```

```python
# 실제 데이터를 막대 그래프로 표시하기

base_url = ("https://raw.githubusercontent.com/practical-jupyter/sample-data/master/anime/")

anime_master_csv = os.path.join(base_url, 'anime_master.csv')

df = pd.read_csv(anime_master_csv, index_col = 'anime_id')

fig = plt.figure()

ax = fig.add_subplot(111)

y = df.groupby('type')['members'].sum()

print(y, '\n')
# print(y.index)
print(type(y))

x = range(len(y))
xlabels = y.index

ax.bar(x, y, tick_label = xlabels)

plt.show()
```
