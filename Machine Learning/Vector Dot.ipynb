* 벡터, 행렬
    - 스칼라 값의 크기
    - 벡터 방향을 가진 값
    - 선형대수학에서는 1차원 행렬을 벡터라고한다. (차원은 벡터의 길이, 요소는 벡터의 차원 ex. array([1, 5, -4]) 는 3차원 벡터 길이는 1)
    - 1차원 배열 : 벡터
    - 2차원 배열 : 행렬
    - 3차원 이상의 배열 : 텐서(Tensor)
* 내적 
    - 방향이 다른 두 벡터의 곱, 두 벡터가 얼마나 비슷한가
    - 내적 결과는 방향이 없다 (스칼라) a dot b = b dot a 
    - 벡터, 행렬, 텐서의 내적(Inner Product, Dot Product)
    - np.dot(a, b)
    - 행렬의 내적 
        - 행, 열 각각의 벡터를 내적한 값을 요소로 가지는 행렬
        - A dot B = C  A(m x n) dot B(n x r) = C(m x r)
        - A의 열과 B의 행이 같아야 한다.
    - 인공지능에서는 내적과 덧셈으로 모든 데이터들의 상관관계를 구한다. 나머지 연산은 의미가 없다.

```python
# 행렬 내적
arr1 = np.array([[1, 2],  # 2차원 벡터가 3개 있다 혹은 3차원 벡터가 2개 있다 (전치)
                 [3, 4],
                 [5, 6]])

arr2 = np.array([[6, 23],
                 [-1, 7],
                 [8, 9]])

# 행렬의 내적
# 3,2 행렬 2개를 내적, 전치해서 앞의 열, 뒤의 행 값을 맞춰 준다.

print(arr2.T)
print(np.dot(arr1, arr2.T))

print(arr1.T)
print(np.dot(arr1.T, arr2))
```

```python
# scikiLearn
from sklearn.datasets import load_iris

iris = load_iris() # 붓꽃 데이터, 품좀마다 차이 (꽃받침 길이, 폭 등)를 모아놓은 데이터

print(iris) # dictionary 키와 밸류  data : 꽃의 길이, target : 꽃의 품종(숫자), target_name : 품종명, ...

# 인공지능에서는 data를 x target을 y로 표현

iris['data'] # iris.data
iris.target # iris['target']

X = iris.data
Y = iris.target
```

```python
print(iris.target_names)
Y == 0 
# boolean indexing
X[Y == 0]  # setosa
```

```python
# 데이터의 통계

X[Y == 0].mean()
print('평균 :', X[Y == 0].mean(axis = 0)) # Y == 0 'setosa'들의 각 data 값의 평균값
print('표준편차 :', X[Y == 0].std(axis = 0)) # 표준편차

for i in range(3) :    # 각 품종의 평균, 표준편차 
    print('이름 :', iris.target_names[i])
    print('평균 :', X[Y == i].mean(axis = 0)) 
    print('표준편차 :', X[Y == i].std(axis = 0)) 
    
# 새로운 붓꽃의 데이터를 구한다면 각 품종의 평균 값과 비교하여 가장 비슷한 품종으로 포함 등등 활용
# 데이터의 값과 평균 값의 거리, 차이 만으로 판단하기는 어렵다. 그래서 딥러닝
```

```python
# 값을 뺀 cos 값만으로 비교, cos 유사도
# a dot b = |a||b|cos0 , cos0 = a dot b / |a||b|
# np.linalg.norm() 벡터의 크기
# 가장 간단한 매칭 모델, 수리모델
# 트위터, 인스타등 추천 알고리즘에 쓰인다

def cos_sim(a, b) :
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))


sample1 = X[Y == 0][0]
setosa = X[Y == 0].mean(axis = 0)
print(sample1) # Y == 0 품종 중 첫번째 데이터와 Y == 0 품종의 평균값 사이의 유사도
print(setosa)
print(cos_sim(sample1, setosa))

sample2 = X[Y == 2][0]
print(cos_sim(sample2, setosa))
```

```python
# 행열 배열 선형대수 병렬처리가 가능하다.
# 샘플 값, 벡터 값 1:1 비교가 아닌 배열로 비교가 가능하고 빅데이터에 사용된다.

# 150개 꽃 샘플, 4가지 값, 4가지 종 
# (150, 4) dot (4, 1) -> (150, 1) 
# 벡터에 한해서 1, 4 (스칼라가 4개) 는 4, 1 로 볼 수 있다.
print(np.dot(X, setosa))
print(np.dot(X, setosa).shape)
print(np.linalg.norm(X, axis = 1))
print(np.linalg.norm(X, axis = 1).shape)
```

```python
# sample 150개에 대해 cos 유사도로 종을 파악하는 코드

def cos_sim2(sample, standard) :   # cos 유사도를 구하는 식
    return np.dot(sample, standard) / (np.linalg.norm(sample, axis = 1) * np.linalg.norm(standard))

sample = X # 150개의 샘플 
setosa = X[Y == 0].mean(axis = 0) # setosa 종의 특성 평균
versicolor = X[Y == 1].mean(axis = 0) # versicolor 종의 특성 평균
virginica = X[Y == 2].mean(axis = 0) # virginica 종의 특성 평균

set_result = cos_sim2(X, setosa)   # 각 샘플과 각 종의 cos 유사도
ver_result = cos_sim2(X, versicolor)
vir_result = cos_sim2(X, virginica)

result = np.stack([set_result, ver_result, vir_result]) # 세 종의 cos 유사도를 한 배열로 결합  stack 이용
# result2 = np.concatenate([set_result, ver_result, vir_result], axis = 1) # (150,3) 의 배열로 합치기 위해서 expand dims을 써야한다 1차원 배열이라서

#np.concatenate((set_result, ver_result), axis = 1)
#print(np.concatenate((set_result, ver_result), axis = 1))

print(np.argmax(result, axis = 0))   # 배열에서 축 기준으로 가장 큰 값의 index 값 반환
print(iris.target) # iris 150개의 샘플의 종 (정답)
print(np.argmax(result, axis = 0) == Y) # 정답 비교 
print(np.sum((iris.target) == np.argmax(result, axis = 0)) / len(Y))   # 정답과 결과(index값)을 비교한 합을 개수로 나눈 정답률

```

* NumPy 연립방정식 풀기 가능, 
* 파이썬의 모든 객체는 얕은 복사, 깊은 복사가 있다. 
    - view, copy

```python
# view (얕은 복사)
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = arr1[1:3] # 슬라이싱할 때 새로운 배열이 생성되는 것이아닌 arr1을 참조하고 있다. arr2는 메모리를 차지하지 않으며 arr1을 보고있다 (view)
print(arr2)
arr2 *= 10  # arr2를 변화하든, arr1을 변화하든 둘 다 영향을 받는다.
print(arr2)
print(arr1)

# 데이터 변환할 때 실수를 해서 원본도 변화 될 수있다. (처음부터 다시 불러와야 된다.)
# 얕은 복사와 깊은 복사는 사용하는 함수에 따라 다르다.
# slicing, indexing, reshape, flatten 은 얕은 복사다.
# 깊은 복사는 대부분의 numpy 연산을 한 이후이다.(일반적으로 연산은 깊은 복사)

# 변화를 원치 않는 경우

arr3 = np.array([1, 2, 3, 4, 5])
arr4 = arr3[1:3].copy()    # .copy() 를 이용하면 view가아닌 새로운 배열을 생성한다.
arr4 *= 10
print(arr4)
print(arr3)
```
