```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np

breast = load_breast_cancer()
breast_df = pd.DataFrame(breast.data, columns = breast.feature_names)
breast_df['diagnosis'] = breast.target

breast_df
```

```python
breast_df.iloc[:, :-1]
```

```python
from sklearn.model_selection import train_test_split

X = breast_df.iloc[:, :-1].values # dataframe에서 마지막열(결과)를 제외하고 numpy array 형태로 가져온다.
y = breast_df.iloc[:, -1].values # 결과값을 numpy array 형태로 가져온다.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = 1) # 20%를 test data로 y를 참고해서 비율을 맞게 data를 random하게 나눈다.

# test로 값을 보고 하이퍼파라미터 튜닝을 반복하는 것은 test값이 간접적으로 머신의 학습에 관여를 한것이다. 실제로는 train을 한번 더 분할해준다.
# validation set을 만들어서 검증하는데 사용한다.

# 이후에 train으로 fit -> Scaler val과 test는 transform, model을 train으로 fit, val으로 모델선택, 하이퍼파라미터 튜닝을 반복하며 가장 좋은 모델을 찾는다.
# 마지막에 test로 최종 테스트, test는 하이퍼파라미터 튜닝에 쓰이면 안되고 마지막 모델 최종 평가에만 쓴다.

# 심화 k겹 교차검증 (k는 train 분할 개수)
# train과 validation을 8:2로 나눌 때 특정 모델에 유리하게 validation값이 정해질수도있다.
# 그래서 train을 5등분하고 1등분을 validation으로 4등분을 train으로 쓰는데 5조각을 validation으로 바꿔가면서 검증하고 평균을 내서 모델을 선택한다.
```

```python
# k겹 교차검증 예제

from sklearn.model_selection import train_test_split, StratifiedKFold

X = breast_df.iloc[:, :-1].values
y = breast_df.iloc[:, -1].values 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = 1) 
```

```python
# kfold # 생성자, 메모리 관리를위해 미리 분할, 생성해서 data를 가지지않고 어떻게 분할할 것인지 정보를 가진 생성자로 나온다

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression # 이름은 회귀인데 분류에 쓴다.

kfold = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 1).split(X_train, y_train) 

#10등분, shuffle하지 않으면 stratify가 작동하지 않는다(순서대로 나눠버리기 때문)분할하는 기계생성
# k = kfold.split(X_train, y_train) 재활용하지 않아서 분할기계 생성하고 바로 스플릿

cvscore = [] # 점수 담는 리스트

for i, (train, val) in enumerate(kfold) : # train, val 셋의 index들 , 두번 unpacking하기 때문에 괄호를 해준다.
    scaler = StandardScaler() 
    X_train_sc = scaler.fit_transform(X_train[train]) # train을 fit하교 scaling, val을 scaling
    X_val_sc = scaler.transform(X_train[val])
    
    pca = PCA(n_components = 2)
    X_train_sc_pca = pca.fit_transform(X_train_sc) # train을 fit하고 특성 뽑기, val을 특성 뽑기
    X_val_sc_pca = pca.transform(X_val_sc)
    
    model = LogisticRegression(solver = 'liblinear', random_state = 1) # 모델에 transform한 train 훈련
    model.fit(X_train_sc_pca, y_train[train])
    # p_val = model.predict(X_val_sc_pca)
    
    score = model.score(X_val_sc_pca, y_train[val]) # 모델 자체에 accuracy score predict도하고 score도 계산한다.
    cvscore.append(score)
    print(f'폴드{i+1}, 정확도:{score}')
    
print('\nCV 정확도:', np.mean(cvscore))


# 교차검증 정확도
```

```python
# 위 코드가 복잡하다면

from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline

lr_pipe = make_pipeline(StandardScaler(), PCA(n_components = 2), LogisticRegression(solver = 'liblinear', random_state = 1)) # 인자로 파이프에서 해야할 일을 순서대로 넣어준다.

# lr_pipe.fit(X_train, y_train) 알아서 pipe라인이 scalring..
# lr_pipe.predict(X_test) 알아서 transform fit..

cvscore = cross_val_score(estimator = lr_pipe, X = X_train, y = y_train, cv = 10, n_jobs = -1) #, scoring = mean_squred_error) # esetimator 추정기 모델로 cv 조각개수만큼 나누고 훈련, 평가 진행
# 파이프라인 : 한쪽입구에 data를 넣으면 파이프를따라 흘러가면서 중간에 scaling, 차원추출등을 한다. 일련의 과정이 자동화된것. 하나의 함수로 모든 과정이 이루어지고 결과만 나온다.

print('\nCV 정확도:', np.mean(cvscore))
```

```python
# 모델의 과대적합을 수치로 판별

# 데이터의 수가 늘어나면 train 데이터의 score는 점점 감소하다가 일정한 점수에 수렴한다. val 데이터의 score는 점점 증가하다가 일정한 점수에 수렴한다.
# 이 그래프를 학습곡선, learning curve라고한다.
# 만약 우리가가진 데이터 10000개로 그렸는데 학습곡선과같이 나온다면 데이터가 특성을 대표할 정도로 충분하다.
# 만약 train data와 val data의 score가 점점 가까워진다면 데이터가 모자라거나, 충분하지 못하다는 의미
# 더 이상가까워지지 않는 train data의 score와 val data의 score 간의 간격이 너무 멀다면 과대적합, 적절하다면 충분히 학습된 것

# 그래프 그려보기
from sklearn.model_selection import learning_curve

lr_pipe = make_pipeline(StandardScaler(), LogisticRegression(solver = 'liblinear', penalty = 'l2', random_state = 1)) #, PCA(n_components = 2)))

sizes, train_score, val_score = learning_curve(estimator = lr_pipe, X = X_train, y = y_train, cv = 10, train_sizes = np.linspace(0.1, 1, 10), n_jobs = -1) # 0 < train_sizes < 1 x축
```

```python
import matplotlib.pyplot as plt

train_mean = np.mean(train_score, axis = 1)
val_mean = np.mean(val_score, axis = 1)
train_std = np.std(train_score, axis = 1)
val_std = np.std(val_score, axis = 1)

plt.figure()
plt.plot(sizes, train_mean)
plt.fill_between(sizes, train_mean + train_std, train_mean - train_std, alpha = 0.2)

plt.plot(sizes, val_mean)
plt.fill_between(sizes, val_mean + val_std, val_mean - val_std, alpha = 0.2)
plt.show()
```

```python
# 모든 파라미터에대해 다 해보기는 어렵다.
# 단순히 모든 경우에 수에대해 반복하며 최적의 값을 찾는 것을 Grid search 방식, 완전 램덤한 random search 방식 등등이 있다.

from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

svc_pipe = make_pipeline(StandardScaler(), SVC(random_state = 1)) # SVC의 고정인자를 제외한 인자는 grid search할 때 넣어준다.

C_range = np.logspace(-4, 3, 8)
gamma_range = np.logspace(-4, 3, 8)
kernel_list = ['linear', 'rbf'] #, 'poly', 'sigmoid', 'precomputed'] # kernel이 linear일 때는 gamma가 필요없다. 오류는 없지만 8번의 낭비한다. 

# 이럴때는 param_grid 딕셔너리를 분리해준다.
param_grid = [{'svc__C' : C_range, 'svc__kernel' : ['linear']}, {'svc__C' : C_range, 'svc__gamma' : gamma_range, 'svc__kernel' : ['rbf']}] # SVC의 인자임을 알려주기위해 '소문자__인자' 로 명시

gs = GridSearchCV(estimator = svc_pipe, param_grid = param_grid, scoring = 'accuracy', cv = 10, n_jobs = -1)
```

```python
gs = gs.fit(X_train, y_train) # gs는 학습결과를 덮어써야한다.
```
