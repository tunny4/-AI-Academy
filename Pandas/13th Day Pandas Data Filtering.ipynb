```python
# Pandas | Data Filtering
import pandas as pd
```

```python
# 불리언 필터링

data = {'이름' : ['아리', '노아', '미아', '리오'],
        '나이' : [3, 5, 4, 2],
        '성별' : ['여자', '남자', '여자', '남자'],
        '키' : [110, 120, 115, 105]}

df = pd.DataFrame(data)

filtered_df = df[df['나이'] > 4]
filtered_df
```

```python
filtered_df = df['나이'] > 4
filtered_df
```

```python
# 나이가 3보다 크고, 성별이 여자 또는 키가 110보다 작다. (조건문 결합)

filtered_df = df[((df['나이'] > 3) & (df['성별'] == '여자')) | (df['키'] < 110)]
filtered_df
```

```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df['target'] = iris.target

print(iris)
```

```python
# 꽃 품종이 versicolor 행들만 선택

iris_df[iris_df['target'] == 1]

# 꽃 받침 길이가 6보다 크다 & 꽃잎 너비가 1보다 작다 & 품종이 virginica이다

iris_df[(iris_df['sepal length (cm)'] > 6) & (iris_df['petal width (cm)'] > 1) & (iris_df['target'] == 2)]
```

```python
# query() 쿼리함수 파이썬에는 없지만 판다스에 있다 SQL 명령문
# query 문장은 " "로 둘러싸야 한다.
# ==, >, <, >=, <=, != 등 비교연산자 활용
# 여러 조건을 결합하기 위한 and(&), or(|) 등을 사용함

data = {'이름' : ['아리', '노아', '미아', '리오'],
        '나이' : [3, 5, 4, 2],
        '성별' : ['여자', '남자', '여자', '남자'],
        '키' : [110, 120, 115, 105]}

df = pd.DataFrame(data)
df
```

```python
filtered_df_01 = df.query("이름 == '아리'")

print(filtered_df_01, '\n')

filtered_df_02 = df.query("키 >= 110")

print(filtered_df_02, '\n')

filtered_df_03 = df.query("키 >= 110 & 이름 == '아리'")

print(filtered_df_03, '\n')

filtered_df_04 = df.query("키 >= 110 or 이름 == '아리'")

print(filtered_df_04, '\n')
```
