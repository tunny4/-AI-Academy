```python
# 슬라이싱 연습
import pandas as pd
import numpy as pn
from sklearn.datasets import fetch_openml

boston = fetch_openml(name = 'house_prices', as_frame = True)
boston_df = pd.DataFrame(data = boston.data, columns = boston.feature_names)
boston_df['target'] = boston.target

boston_df
```

```python
# 1. 데이터프레임의 처음 5개, 마지막 5개 행 출력

print(boston_df.head())
print(boston_df.tail())

# 2. 인덱스 10부터 15까지 행 출력

print(boston_df.iloc[10:16])

# 3. Id, MSSubClass, MSZoning 열의 데이터만 선택 출력

print(boston_df.loc[:, ['Id', 'MSSubClass', 'MSZoning']])

# 4. OverallQual, OverallCond, GarageCars 열의 데이터만 선택하여 출력

print(boston_df.loc[:, ['OverallQual', 'OverallCond', 'GarageCars']])

# 5. SaleType, SaleCondition, target 열에 있는 인덱스 라벨 100부터 110까지 행 출력

print(boston_df.loc[100:110, ['SaleType', 'SaleCondition', 'target']])

# 6. MSSubClass, LotArea 열에 있는 인덱스 라벨 50부터 60까지 행 출력

print(boston_df.loc[50:60, ['MSSubClass', 'LotArea']])

# 7. 전체 프레임 데이터에서 LotArea 열의 데이터가 15000보다 작은 행들만 출력

print(boston_df[boston_df['LotArea'] < 15000])

# 8. 전체 데이터 프레임에서 LotFrontage 열의 데이터가 80 이상, SaleCondition 열의 데이터가 Normal인 행들만 선택 출력 (두 조건식의 결헙하는 논리 연산자 &)

print(boston_df[(boston_df['LotFrontage'] > 80) & (boston_df['SaleCondition'] == 'Normal')])
```
