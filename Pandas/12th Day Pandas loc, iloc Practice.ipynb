```python
# 실습
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame([[0.3, 0.2, 0.6, 0.9], 
                   [0.7, 0.3, 0.3, 0.6],
                   [0.1, 0.5, 0.7, 0.5],
                   [0.6, 0.4, 0.1, 0.3],
                   [0.8, 0.9, 0.2, 0.4]], columns = ['A', 'B', 'C', 'D'])

df
```

```python
# 1번 loc iloc 

print("C 열의 2 행 출력 :", df.loc[2]['C'], df.loc[2, 'C'])
print("2 열의 3 행 출력 :", df.iloc[2]['B'])
# df['B'].loc[1:3]
print("B ~ D 열의 1 ~ 3 행 출력 \n", df.loc[[0, 1, 2], ['B', 'C', 'D']], '\n', df.loc[0:2, 'B':'D'])
# df['A'].iloc[2:5]
print("0 ~ 2 열의 2 ~ 4 행 출력 \n", df.iloc[2:5, 0:3])
```

```python
# 2번 

from sklearn.datasets import load_iris

iris = load_iris() # iris 데이터셋 로드

iris_df = pd.DataFrame(data = iris.data, columns = iris.feature_names) # 데이터 프레임 생성

iris_df['target'] = iris.target # target 열 추가

iris_df
```

```python
# 1.sepal length 열의 모든 값
iris_df.loc[:, 'sepal length (cm)']

# 2.인덱스가 2인 행
iris_df.iloc[2]

# 3.인덱스 라벨이 2인 행
iris_df.loc[2]

# 4.인덱스가 3인 행부터 인덱스가 7인 행까지
iris_df.iloc[3:8]

# 5.인덱스 라벨이 3인 행부터 인덱스 라벨이 7인 행까지
iris_df.loc[3:7]

# 6. 인덱스 라벨이 5인 행의 petal length 열의 값
iris_df.loc[5, 'petal length (cm)']

# 7.전체 데이터 프레임에서 sepal width 열에서 3보다 큰 행만 선택
iris_df.loc[3:, 'sepal width (cm)']

for i in iris_df.index :
    if iris_df.loc[i, 'sepal width (cm)'] > 3 :
        print("{} : {}" .format(i,iris_df.loc[i, 'sepal width (cm)']))
    
iris_df[iris_df['sepal width (cm)'] > 3]
```
