```python
# 롤링 윈도우 (Rolling Window) 
import pandas as pd
data = {'date' : pd.date_range(start = '2023-05-01', end = '2023-05-10', freq = 'D'),
        'value' : [1, 6, 4, 12, 36, 17, 44, 91, 180, 23]}

df = pd.DataFrame(data)

df.set_index('date', inplace = True)

df
```

```python
rolling_mean = df['value'].rolling(window = 5).mean()  # 이동 평균 함수 1개씩 이동하면서 5개의 값의 평균을 구한다. 1, 2, 3, 4는 값이 5개가 되지않아 NaN
# 필요없는 값을 버릴 때 이용 ???

rolling_mean
```

```python
rolling_mean = df['value'].rolling(window = 5)

for i in rolling_mean :
    print(i)
    
# [print(i) for i in rolling_mean]
```
