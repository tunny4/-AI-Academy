```python
# 두 개의 Y축을 가진 그래프 (양쪽으로 Y축을 다르게 쓰는 그래프)
import os
import pandas as pd
import matplotlib.pyplot as plt

if os.name == 'posix' :
    plt.rc('font', family = 'AppleGothic')
else :
    plt.rc('font', family = 'Malgun Gothic')
    
base_url = ("https://raw.githubusercontent.com/practical-jupyter/sample-data/master/anime/")

t4816_csv = os.path.join(base_url, "4816.csv")

df = pd.read_csv(t4816_csv, index_col = 0)

df.index = pd.to_datetime(df.index)

df
```

```python
# bar : 막대 그래프, plot : 꺾은 선 그래프 

fig = plt.figure(figsize = (16, 6))

ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()     # twinx 함수 

ax1.plot(df.index, df['Close'], label = 'Close', color = 'red')
ax2.bar(df.index, df['Volume'], label = 'Volume', color = 'green') 

ax1.set_yticks([i * 2000 for i in range(5)])
ax2.set_yticks([i * 50000 for i in range(5)])

ax1.legend(loc = 2)
ax2.legend(loc = 1)
plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

if os.name == 'posix' :
    plt.rc('font', family = 'AppleGothic')
else :
    plt.rc('font', family = 'Malgun Gothic')
```

```python
# 산포도 
# 산포도는 x축과 y축에 수량이나 크기 등을 대응시켜서 적합한 점에 데이터를 표현

np.random.seed(0)

x = np.random.rand(100)
y = np.random.rand(100)

fig = plt.figure(figsize = (3, 3))
ax = fig.add_subplot(111)

ax.scatter(x, y, color = 'indigo')

plt.show()
```

```python
base_url = ("https://raw.githubusercontent.com/practical-jupyter/sample-data/master/anime/")

anime_master_csv = os.path.join(base_url, 'anime_master.csv')

df = pd.read_csv(anime_master_csv, index_col = 'anime_id')

df
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(df['members'], df['rating'], alpha = 0.5, color = 'royalblue')   # alpha : 투명도

plt.show()
```

```python
types = df['type'].unique()

fig = plt.figure()

ax = fig.add_subplot(111)

for t in types : 
    x = df.loc[df['type'] == t, 'members']
    y = df.loc[df['type'] == t, 'rating']
    ax.scatter(x, y, alpha = 0.5, label = t)
    
plt.show()
```

```python
# 특징을 이용해 시각화. 각 축에 어떤 값을 너을지 고민, 상관 관계 분석으로 비슷한 2개를 이용

from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris = load_iris()

data = iris.data
target = iris.target
feature_names = iris.feature_names

# 분류 모델 학습 지도학습 : 정답을 알려주면서, 비지도학습 : 정답없이 안 알려주는 학습
# target이 정답

fig = plt.figure(figsize = (16, 8))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

colors = np.arange(150)

ax1.scatter(data[:, 0], data[:, 1], c = colors)
ax1.set_xlabel(feature_names[0])
ax1.set_ylabel(feature_names[1])
ax1.set_title('Sepal Length VS Sepal Width')

ax2.scatter(data[:, 2], data[:, 3], c = colors)
ax2.set_xlabel(feature_names[2])
ax2.set_ylabel(feature_names[3])
ax2.set_title('Petal Lengh VS Petal Width')

plt.show()
```

```python
# 히스토그램

mu = 100
sigma = 10

np.random.seed(0) # 0을 중심으로 random값 생성하기 위해 중심설정

x = np.random.normal(mu, sigma, 10000) # 정규 분포도에 따라 랜덤값 생성

fig = plt.figure()
ax = fig.add_subplot(111)

ax.hist(x, rwidth = 0.9, bins = 20) # 히스토그램 rwidth 바의 폭, bins 그래프의 개수

plt.show()
```

```python
base_url = ("https://raw.githubusercontent.com/practical-jupyter/sample-data/master/anime/")

anime_master_csv = os.path.join(base_url, 'anime_master.csv')

df = pd.read_csv(anime_master_csv, index_col = 'anime_id')

df_tv = df[df['type'] == 'TV']

df_tv
```
