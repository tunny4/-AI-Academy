# OpenCV 프로그래밍 라이브러리

* 영상처리와 컴퓨터 비전 분야의 대표적인 라이브러리
* 인공지능 모델 신생아 -> 충분한 학습 데이터가 필요
* 정상 데이터는 많지만, 비정상 데이터는 충분치 않다.
* 분류 모델은 정상 데이터와 비정상 데이터의 비율이 같으면 좋다.
* 비율이 달라도 그 수가 충분히 많으면 괜찮다.
* 데이터 수가 충분치 않으면 가짜 데이터를 만들어서 학습시킨다. (특히 의료분야)
    - 원본 사진을 크롭, 회전, 노이즈 등등..
* 데이터의 불균형을 해결하는 방법 중 하나가 데이터 증강이다. (가짜 데이터 : 실제 근거에 의해 만든 데이터)
* 데이터 증강은 학습 데이터를 늘이는 방법을 총칭한다.
* 데이터 증강을 통해 많은 학습 데이터를 확보할 수 있다. 하지만 실제 데이터를 많이 수집하는 것이 좋다.

```python
import cv2
```

```python
# 중요한 코드 

image = cv2.imread("image.png", cv2.IMREAD_COLOR) #image read / IMREAD_COLOR = rgb, IMREAD_GRAYSCALE = 흑백, IMREAD_REDUCED_GRAYSCALE_2,4,8 = 작은 사진  / 절대경로, 상대경로(파일과 같은 위치) 
print(image.shape)  # 이미지는 높이, 너비, 채널 수로 3차원이다.

height, width, channel = image.shape # 튜플 언패킹 
print(height, width, channel)

cv2.imshow("numpy", image) # 이미지 출력, 3줄은 같이 쓰이는 패턴 
cv2.waitKey() # 키보드를 누르면 종료된다.
cv2.destroyAllWindows() 

# 픽셀기반 이미지, 레스터 이미지(jpg, png, gif 등)는 openCV에서 처리가 가능 벡터기반 이미지 파일(pdf, svg) 등은 처리가 불가능하다.
```

```python
image = cv2.imread("image.jpg", cv2.IMREAD_COLOR) #image read / IMREAD_COLOR = rgb, IMREAD_GRAYSCALE = 흑백, IMREAD_REDUCED_GRAYSCALE_2,4,8 = 작은 사진  / 절대경로, 상대경로(파일과 같은 위치) 
print(image.shape)
```

```python
height, width, channel = image.shape # 튜플 언패킹 
print(height, width, channel)
```

```python
fliped_image = cv2.flip(image, -1) # src 이미지를 0 상하 반전, 1 좌우 반전, -1 상하 좌우 반전 
```

```python
rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), -90, 1)  # 회전 행렬
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height), borderMode = cv2.BORDER_REPLICATE) # 왜곡시키다, borderMode 빈 공간을 채워줌 REFLECT
```

```python
upSampled_image = cv2.pyrUp(image, dstsize = (width * 2, height * 2), borderType = cv2.BORDER_DEFAULT) # 이미지 scale 변경 dstsize로 조절 pyr 피라미드드
downSampled_image = cv2.pyrDown(image, dstsize = (width // 2, height // 2))
```

```python
# resizing
absolute_image = cv2.resize(image, dsize = (640, 480), interpolation = cv2.INTER_LINEAR) # 이미지 사이즈를 직접 변경 이미지 크기 지정
relative_image = cv2.resize(image, dsize = (0, 0), fx = 0.3, fy = 0.7, interpolation = cv2.INTER_AREA) # x축 y축 크기 지정
```

```python
cv2.imshow("Original Image", image)
# cv2.imshow("Fliped Image", fliped_image)  # 반전
# cv2.imshow("Rotation Image", rotated_image) # 회전
# cv2.imshow("Upsampled Image", upSampled_image) # 확대
# cv2.imshow("Downsampled Image", downSampled_image) # 축소
cv2.imshow("Absolute Image", absolute_image)
cv2.imshow("Relative Image", relative_image)
cv2.waitKey()
cv2.destroyAllWindows() 
```

```python
# 오후 수업
# 이 화면에 출력하는게 보기 편하기 때문에 matplotlib 을 쓴다.
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# RGB, BGR 같이 채널 표준이 다르기 때문에 색이 다르게 나온다.
# openCV에 channel convert를 이용 BGR, RGB를 바꾼다.

img = cv2.imread("image.jpg")   

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # numpy에서 채널의 순서를 바꿔준다. 
# img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # 흑백 
print(img)
plt.imshow(img)
# plt.imshow(img, cmap='gray') # 흑백, 3채널이면 plt가 알아서 출력하지만 1채널이면 cmap으로 color mapping이 필요하다.

plt.show()
```

```python
# 이미지 크기 자르기 슬라이싱으로 자른다.
#crop_img = img[40:140, 25:125] # 채널은 모두 다 가져와야한다.     # 얕은 복사라 원본에 영향을 미친다.
crop_img = img[40:140, 25:125].copy() # 깊은 복사라 원본에 영향을 안미친다.
crop_img[40:60] = cv2.cvtColor(crop_img[40:60], cv2.COLOR_RGB2BGR) # view의 이미지에 영향을 주면, 얕은 복사라 원본에 영향을 준다.

plt.imshow(crop_img)
plt.show()

plt.imshow(img)
plt.show()
```

```python
# 역상 (색상 반전 )
# RGB 0 ~ 255  0을 255로, 100을 155로 바꾸는 것

bn_img = cv2.bitwise_not(img) # 비트와이즈 낫 (원본)
plt.imshow(bn_img)
plt.show()
print(img)
print(bn_img)
```

```python
# 2진화 : 인공지능 성능이 낮을 때, 컬러풀한 이미지 처리를 잘 못할 때 썼다. 요즘은 잘 쓰지 않는다.
# 0 또는 1로 RGB 값을 바꾼다. 판화처럼 변한다.
# 우선 채널을 합친다.
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
print(gray_img.shape)

_, bin_img = cv2.threshold(gray_img, 122, 255, cv2.THRESH_BINARY) # threshold = 경계값, 임계값 / 나눌 기준값, ? / 계산된 값(설정한 임계값), 결과 2가지를 반환한다. / 안쓰는 값은 _ 로 받는다.

print(_)
plt.imshow(bin_img, cmap = 'gray')
plt.show()

print("흑백 :", gray_img)
print("2진 :", bin_img)
```

```python
# 블러 처리, 네모난 필터가 이미지를 지나가면서 평균을 내고 그 값으로 변환시킨다. 
# 잡음 제거 용도로 쓰인다.
blur_img = cv2.blur(img, (16, 16)) # 필터 사이즈 지정, 2의 n승으로 해야 표준해상도 일때 딱 맞다
plt.imshow(blur_img)
plt.show()
```

```python
# 윤곽선 강조, 따기
# 인공지능이 사물을 명확하게 인식하도록

sobel_img = cv2.Sobel(gray_img, cv2.CV_8U, 1, 0, 5) # cv2.CV_8U 가장자리 설정할 때 뎁스를 설정해 주는 것 
                                        # 배경 -> 사물, 사물 -> 배경 으로 넘어갈 때 값의 급격한 변화를 보고 경계로 판단한다. 미분 차수, 커널 홀수만 최대 31 커널이 커지면 날카로운 부분, 작으면 ...
plt.imshow(sobel_img, cmap = 'gray')
plt.show()
    
```

```python
# 윤곽선 강조, 따기
# 인공지능이 사물을 명확하게 인식하도록

laplac_img = cv2.Laplacian(gray_img, cv2.CV_8U, ksize = 5) # cv2.CV_8U 가장자리 설정할 때 뎁스를 설정해 주는 것 
                                        # 배경 -> 사물, 사물 -> 배경 으로 넘어갈 때 값의 급격한 변화를 보고 경계로 판단한다. 미분 차수, 커널 최대 31 커널이 커지면 날카로운 부분, 작으면 ...
plt.imshow(laplac_img, cmap = 'gray')
plt.show()
```

```python
# 라플라시안 개선해서 나옴, 좀 더 다양한 상황에서 사용 가능

canny_img = cv2.Canny(gray_img, 122, 255)
plt.imshow(canny_img, cmap = 'gray')
plt.show()
```

```python
# 색상 검출
# hsv 색상표 Hue (각도 = 색상) Saturation (원심 거리 = 채도) Value (높이 = 명도)
# 색상 검출은 Hue를 필요로한다. 210 240

print(img)
hsv_img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV) # 사진을 hsv 채널로 변경
h, s, v = cv2.split(hsv_img) # 채널 별로 나누기
print(hsv_img)
# print(img)
# print(hsv_img)

plt.imshow(h, cmap = 'hsv')
plt.show()
plt.imshow(s, cmap = 'hsv')
plt.show()
plt.imshow(v, cmap = 'hsv')
plt.show()
print(h)
h_range = cv2.inRange(h, 0, 30) # inRange(채널, 뽑고 싶은 범위)  색 범위 마스크
azure = cv2.bitwise_and(hsv_img, hsv_img, mask = h_range) # true면 뽑고 false면 안 뽑는다.  and 연산 동일한 이미지를 넣어서 자기 자신과 and 연산하게 한다
azure = cv2.cvtColor(azure, cv2.COLOR_HSV2RGB) # HSV채널을 출력하기위해 RGB 채널로 바꾼다.
plt.imshow(azure, cmap ='hsv')
plt.show()
```

```python
hsv_img = cv2.imread("hsv.png")   

hsv_img = cv2.cvtColor(hsv_img, cv2.COLOR_BGR2RGB) # numpy에서 채널의 순서를 바꿔준다. 

plt.imshow(hsv_img)
plt.show()

hsv_img = cv2.cvtColor(hsv_img, cv2.COLOR_RGB2HSV) # 사진을 hsv 채널로 변경
m, _, _ = cv2.split(hsv_img)
m_range = cv2.inRange(m, 0, 30)
m_img = cv2.bitwise_and(hsv_img, hsv_img, mask = m_range)
m_img = cv2.cvtColor(m_img, cv2.COLOR_HSV2RGB)
print(hsv_img)
plt.imshow(hsv_img, cmap = 'hsv')
plt.show()
plt.imshow(m_img)
plt.show()
```
