# 시각화 알고리즘

* seaborn
* matplotlib
    - 제일 기초, 기본. 나머지는 matplot이 부족하거나 예쁘게 만들기위해 사용

        - graph를 그리기 위해서는 figure와 subplot을 먼저 만들어야 된다.

```python
# 라이브러리 호출
import matplotlib.pyplot as plt
```

```python
# figure 생성
fig = plt.figure()

# subplot 생성
ax1 = fig.add_subplot(221) # 2행 2열 1번 위치
ax2 = fig.add_subplot(2, 2, 2) # 2행 2열 2번 위치  i, j, k를 넣고 for문을 돌려 생성하기도 한다.
ax3 = fig.add_subplot(2, 2, 3) # 2행 2열 3번 위치 콤마로 나눠도 똑같다
```

```python
# subplot 생성_2
fig, axes = plt.subplots(2, 2) # axes는 numpy 배열타입

print(type(axes), '\n')
print(axes)

plt.show()
```

```python
fig, axes = plt.subplots(2, 2)

axes[0, 0].set_title("Subplot 0-0")
axes[0, 1].set_title("Subplot 0-1") # 0행 1열 subplot에 제목 지정

plt.show()
```

```python
# 2행 5열의 subplot을 만들고 각 그림에 제목을 붙여 봅시다.
fig, axes = plt.subplots(2, 5, figsize = (12, 8)) # figsize : figure 크기 조절

title_list = [range(1, 6), range(6, 11)]

for i in range(2) :
    for j in range(5) :
        
        # axes[i, j].set_title("Subplot {}-{}" .format(i+1, j+1))
        axes[i, j].set_title("Subplot {}" .format(title_list[i][j]))

plt.show()
```

```python
plt.style.use('dark_background') # 스타일 변경 ggplot, classic, bmh, seaborn, grayscale, fivethirtyeight, dark_background

fig = plt.figure(figsize = (2, 2)) # figure size 조절

ax = fig.add_subplot(111)

plt.show()
```

```python
# 그림그리기

plt.style.use('dark_background') # 스타일 변경 ggplot, classic, bmh, seaborn, grayscale, fivethirtyeight, dark_background

fig = plt.figure(figsize = (2, 2)) # figure size 조절

ax = fig.add_subplot(111)

data = [0, 1]

ax.plot(data)  # plot 직선그리는 함수

plt.show()
```

```python
# Matplotlib 그래프 관련 주요 함수

plt.style.use('dark_background') # 스타일 변경 ggplot, classic, bmh, seaborn, grayscale, fivethirtyeight, dark_background

fig = plt.figure(figsize = (8, 2)) # figure size 조절 함수

ax = fig.add_subplot(111)

data = [0, 1, 3, 20, 10, 30]

# plot 직선그리는 함수
ax.plot(data)  

# set_title 제목 지정 함수, fontdict 인자 font 관련 옵션 딕셔너리 형태, loc 인자 제목 위치 지정
ax.set_title("My First Graph", fontdict = {'fontsize' : 30, 'fontweight' : 'bold'}, loc = 'left', color = 'green') 

# xlabel, ylabel x축 y축에 제목 지정 함수, set_title과 옵션을 같이 쓰는게 많다.
ax.set_xlabel('X Axis', fontdict = {'fontsize' : 10, 'fontweight' : 'bold'}, loc = 'right', color = 'yellow')
ax.set_ylabel('Y Axis', fontdict = {'fontsize' : 10, 'fontweight' : 'bold'}, loc = 'top', color = 'red') # y축 loc 인자 값이 top, center, bottom

# tick 눈금 설정 함수
ax.set_xticks([1, 3, 5])
ax.set_yticks([10, 20, 30])

# xticklabels, yticklabels 숫자로 지정된 눈금 값을 문자로 지정하는 함수
ax.set_xticklabels(['A', 'B', 'C'])
ax.set_yticklabels(['D', 'E', 'F'])

plt.show()
```
