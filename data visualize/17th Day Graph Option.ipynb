```python
# 라이브러리 호출
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# figure 생성, subplot 배치, graph 작성

fig = plt.figure()

ax = fig.add_subplot(111)

x = [0, 1]
y = [1, 0]

# 그래프 그리기
# plot bar scatter hist

ax.plot(x, y)

# 옵션 추가

plt.show()
```

```python
# figure 서식 지정

fig = plt.figure(figsize = (8, 8),  # figsize : figure 크기
                dpi = 100, facecolor = 'gainsboro', # dpi : 해상도, facecolor : figure 색채우기
                edgecolor = 'mistyrose', linewidth = 10, frameon = True, # edgecolor : 테두리 색상 ('k'는 검정색), linewidth : 테두리 두께, frameon : 테두리 생성
                tight_layout = False) # subplot의 배치 간격 figsize와 같이 조절해야 된다.

fig.subplots_adjust(wspace = 0.1, hspace = 0.1) # tight_layout이 False 일 때, subplot 간격 조절 ?????????????

# subplot 서식 지정

ax1 = fig.add_subplot(221, facecolor = 'oldlace', # 111 : 1행 1열 1번째, facecolor : subplot 색채우기
                    xlim = [0, 10], ylim = [0, 10000]) # xlim, ylim : limit 값의 범위 지정 x,y limit 설정 함수도 따로 있다.

ax2 = fig.add_subplot(222, facecolor = 'wheat', 
                    xlim = [0, 10], ylim = [0, 10000])

ax3 = fig.add_subplot(223, facecolor = 'tan', 
                    xlim = [0, 10], ylim = [0, 10000])

ax4 = fig.add_subplot(224, facecolor = 'bisque', 
                    xlim = [0, 10], ylim = [0, 10000])


x = [0, 5, 8, 10]
y = [1000, 8000, 3000, 6000]

ax1.plot(x, y)
ax2.plot(x, y)
ax3.plot(x, y)
ax4.plot(x, y)

plt.show()
```

```python
# 선의 굵기와 색 지정

x = np.array([-0.2, 1])
y = [0, 1]

fig = plt.figure(figsize = (9, 3))

ax = fig.add_subplot(111, xlim = (-0.2, 4))

ax.plot(x, y, linewidth = 1, color = 'b', label = "linewidth = 1 (alpha = 1)") # linewidth : 선의 굵기 , alpha : 선의 투명도

ax.plot(x + 0.1, y, linewidth = 2, color = 'g', alpha = 0.5, label = "linewidth = 2 (alpha = 0.5)") # numpy array이기 때문에 연산이 가능하다.

ax.plot(x + 0.2, y, linewidth = 10, color = 'r', alpha = 0.3, label = "linewidth = 10 (alpha = 0.3)") 

plt.legend()
plt.show()
```

```python
# 선의 끝 모양 지정 

x = [0.1, 1]
y = np.array([0, 0])

fig = plt.figure(figsize = (6, 3))

ax = fig.add_subplot(111, xlim = (0, 1.7), ylim = (-0.2, 1.2))

ax.plot(x, y + 1, linewidth = 12, color = 'lightblue', solid_capstyle = 'butt', label = 'butt') # solid_capstyle : 선의 끝 스타일, butt : ???

ax.plot(x, y + 0.5, linewidth = 12, color = 'aqua', solid_capstyle = 'round', label = 'round') # round : 둥글게 (길어져서 정확한 수치를 보기에 부적절하다.)

ax.plot(x, y, linewidth = 12, color = 'limegreen', solid_capstyle = 'projecting', label = 'projecting') # projecting : ???

plt.legend(loc = 0) # loc 문자로 쓸 수도있고 숫자로 대체 가능하다. 0은 스스로 가장 좋은 위치
plt.show()
```

```python
# 선의 결합점 모양 지정 (선이 꺾이는 곳)

x = np.array([0, 1, 2, 3])
y = [0, 1.0, 0.2, 1.2]

fig = plt.figure(figsize = (6, 3))

ax = fig.add_subplot(111, xlim = (0, 4.5), ylim = (0, 1.2))

ax.plot(x, y, linewidth = 10, color = 'lightblue', solid_joinstyle ='miter', label = 'miter')  # solid_joinstyle : 선의 결합점 모양, miter : default

ax.plot(x + 0.5, y, linewidth = 10, color = 'lightskyblue', solid_joinstyle ='round', label = 'round') # round

ax.plot(x + 1, y, linewidth = 10, color = 'skyblue', solid_joinstyle ='bevel', label = 'bevel') # bevel


plt.legend(loc = 0) # loc 문자로 쓸 수도있고 숫자로 대체 가능하다. 0은 스스로 가장 좋은 위치
plt.show()
```

```python
# 선의 스타일 지정

x = np.array([0, 1])
y = np.array([1.9, 1.9])

fig = plt.figure(figsize = (6, 3))

ax = fig.add_subplot(111, xlim = (-0.1, 1.7), ylim = (0, 2))

ax.plot(x, y, linewidth = 5, color = 'lightblue', linestyle = 'solid', label = 'solid') # linestyle : 실선, 점선 등 선의 스타일 지정

ax.plot(x, y - 0.1, linewidth = 5, color = 'lightskyblue', linestyle = '-', label = '-') # 선으로 지정가능 - = solid, -- = dashed, : = dotted

ax.plot(x, y - 0.3, linewidth = 5, color = 'springgreen', linestyle = 'dashed', label = 'dashed') 

ax.plot(x, y - 0.4, linewidth = 5, color = 'seagreen', linestyle = '--', label = '--')

ax.plot(x, y - 0.6, linewidth = 5, color = 'violet', linestyle = 'dotted', label = 'dotted')

ax.plot(x, y - 0.7, linewidth = 5, color = 'blueviolet', linestyle = ':', label = ':')

plt.legend(loc = 0) # loc 문자로 쓸 수도있고 숫자로 대체 가능하다. 0은 스스로 가장 좋은 위치
plt.show()
```

```python
# 산포도에서 점의 색과 불투명도 지정

x = np.arange(0, 6, 0.1)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(x, y, label = 'sin', color = 'lightblue', edgecolor = 'blue') # edgecolor : 점 테두리 색 지정
ax.scatter(x, z, label = 'cos', color = 'springgreen', edgecolor = 'green', alpha = 0.3)

plt.legend(framealpha = 0.1, loc = 'lower right') # framealpha : frame 투명도
plt.show()
```

```python
# 컬러맵 사용하기

import matplotlib.cm as cm

fig = plt.figure()

ax = fig.add_subplot(111)

x = np.arange(20)
y = np.arange(20)

ax.scatter(x, y + 5, color = cm.spring(0.9)) # color map의 spring 중 선택한 색상
ax.scatter(x, y, color = cm.get_cmap('spring')(0.9)) # 위와 같은 결과가 나온다.
ax.scatter(x, y + 10, color = cm.plasma(0.5))
ax.scatter(x, y + 15, color = cm.tab10(0.5))

plt.show()
```

```python
# 텍스트 그리기

fig = plt.figure()

ax = fig.add_subplot(111)

fps = {'family' : 'fantasy', 'style' : 'italic', 'weight' : 'heavy', 'size' : 40, 'color' : 'Green'}

ax.text(0, 8, 'Text', alpha = 0.5, rotation = 0, # size = 20, color = 'red', # text : 글자를 그리는 함수 x, y 좌표, text, size, color, alpha, rotation 등 인자가 있다.
        fontdict = fps ) # font 설정 fontdict

ax.set_xticks([i * 2 for i in range(6)])
ax.set_yticks([i * 2 for i in range(6)])

plt.show()
```
