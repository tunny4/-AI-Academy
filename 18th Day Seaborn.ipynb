# seaborn : 데이터 시각화

* matplotlib와 같이 많이 쓰인다. 호환이 됨
* matplotlib보다 세련됨, 다양한 그래프
* 다양한 데이터셋 지원 (load_dataset : 불러오고자하는 데이터셋 이름)
* 불러온 데이터셋은 모두 데이터프레임 형태이다.

```python
# seaborn 라이브러리 불러오기
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid') # whitegrid, dark, white, ticks ...
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.histplot(x = 'age', data = titanic, palette = 'Paired', hue = 'survived')
```

```python
# 산점도 : scatterplot() 함수

tips = sns.load_dataset('tips')

tips # size : 인원수

                                                                       # 파라미터
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, alpha = 0.5, # data : 불러온 dataset, Dataframe 형식이라 쓰기 편하다
                 hue = 'sex', size = 'size', sizes = (40, 200)) # hue : data에 따라 항목 나누기, size : data에 따라 크기 지정, sizes : 튜플 형태로 a ~ b까지 크기를 자동 지정

plt.xlabel('Total Bill')  # seaborn 그래프에 matplotlib 함수도 쓸 수 있다.
plt.ylabel('Tip')
plt.title('Relationship between Total Bill and Tip')

plt.show()
```

```python
# 라인 플롯 : lineplot() 함수, 시간에 따른 변화를 나타 낼 때 

fmri = sns.load_dataset('fmri')

fmri # 자기공명...

sns.lineplot(x = 'timepoint', y = 'signal', data = fmri,
             hue = 'event', style ='region', markers = True ) # style : data에 따라 다른 Style로 그래프가 추가된다. markers : 값에 점을 찍어 준다.

plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.title('Changes in Signal Intensive over Time')
```

```python
# 바 플롯 : barplot() 함수

titanic = sns.load_dataset('titanic')

sns.barplot(x = 'class', y = 'fare', errorbar = None, data = titanic,
            hue = 'sex', palette = 'cool')

plt.title('Average Fare by Class and Gender on the Titanic')
plt.show()
```

```python
# 히스토그램 : histplot() 함수

iris = sns.load_dataset('iris')


sns.histplot(x = 'petal_length', bins = 20, data = iris, color = 'green', kde = True) # kde : 그래프에 맞게 선을 그어주는 인자 bins : 막대 개수,
```

```python
# 커널 밀도 플롯 : kdeplot() 함수, 데이터의 밀도? 양?

tips = sns.load_dataset('tips')

sns.kdeplot(x = 'total_bill', data = tips, hue = 'time', fill = True, palette = 'summer', alpha = 0.3) # fill 인자로 채우기 선택

plt.title('Density Plot of Total Bill by Meal Time')
plt.show()
```

```python
# 박스 플롯 : boxplot() 함수
# 점들 : 이상치

tips = sns.load_dataset('tips')

sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'time', palette = 'autumn', linewidth = 3) 

plt.title('Box Plot of Total Bill by Day and Meal Time')
plt.show()
```

```python
# 히트맵 : heatmap()

tips = sns.load_dataset('tips')

pivot = tips.pivot_table(index = 'day', columns = 'size', values = 'tip')

sns.heatmap(pivot, cmap = 'crest', annot = True) # annot 숫자 값이 나온다.

plt.show()
```

```python
# Pandas 상관관계 corr() 함수 : 변수간의 상관관계

titanic = sns.load_dataset('titanic')

titanic.corr()

sns.heatmap(titanic.corr(), annot = True, cmap = 'summer')

plt.show()
```

```python
# 조인트 플롯 : jointplot() 함수 
# 산점도와 히스토그램을 결합한 플롯

tips = sns.load_dataset('tips')

sns.jointplot(x = 'total_bill', y = 'tip', height = 8, data = tips, kind = 'hex') # kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }

plt.show()
```

```python
# 패시 그리드 : FacetGrid() 함수

tips = sns.load_dataset('tips')

g = sns.FacetGrid(tips, col = 'day')

g.map(sns.histplot, 'total_bill')
```

```python
iris = sns.load_dataset('iris')

g = sns.FacetGrid(iris, col = 'species')

g.map(sns.scatterplot, 'petal_width', 'petal_length')
```
